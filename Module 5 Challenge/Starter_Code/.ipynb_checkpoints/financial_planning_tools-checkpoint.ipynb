#### Review the endpoint URLs for the API calls to Free Crypto API in order to get the current pricing information for both BTC and ETH.

```python
# The Free Crypto API Call endpoint URLs for the held cryptocurrency assets
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=USD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=USD"
```

#### Step 2. Use the Requests library to get the current price (in US dollars) of Bitcoin (BTC) and Ethereum (ETH) by using the API endpoints that the starter code supplied.

```python
# Using the Python requests library, make an API call to access the current price of BTC
btc_response = requests.get(btc_url).json()

# Using the json.dumps function to review the response data from the API call in a user friendly format

print(json.dumps(btc_response, indent=4, sort_keys = True))
```

```python
# Using the Python requests library, making an API call to access the current price of ETH
eth_response = requests.get(eth_url).json()

# Using the json.dumps function to review the ethereum response data from the API call in a user friendly format
print(json.dumps(eth_response, indent=4, sort_keys = True))
```

#### Step 3: Navigate the JSON response object to access the current price of each coin, and store each in a variable.

```python
# Navigating through the BTC response object to access the current price of ETH
eth_price = eth_response["data"]['1027']['quotes']['USD']['price']


# Printing the current price of ETH
print(eth_price)
```

### Step 4: Calculate the value, in US dollars, of the current amount of each cryptocurrency and of the entire cryptocurrency wallet.

```python
# Computing the current value of the BTC holding 
btc_value = btc_price * btc_coins

# Printing the current value of your holding in BTC
print(btc_value)
```

```python
# Computing the current value of the ETH holding 
eth_value = eth_price * eth_coins

# Printing the

#current value of your holding in ETH
print(eth_value)

```

```python
# Computing the total value of the cryptocurrency wallet

total_crypto_wallet = eth_value + btc_value

# Printing current cryptocurrency total wallet balance
print(total_crypto_wallet)
```

In this section, you’ll determine the current value of a member’s stock and bond holdings. You’ll make an API call to Alpaca via the Alpaca SDK to get the current closing prices of the SPDR S&P 500 ETF Trust (ticker: SPY) and of the iShares Core US Aggregate Bond ETF (ticker: AGG). For the prototype, assume that the member holds 110 shares of SPY, which represents the stock portion of their portfolio, and 200 shares of AGG, which represents the bond portion. To do all this, complete the following steps:

4. Get the current closing prices for `SPY` and `AGG` by using the Alpaca `get_barset` function. Format the response as a Pandas DataFrame by including the `df` property at the end of the `get_barset` function.

#### Review the total number of shares held in both (SPY) and (AGG).

```python
# Displaying the current amount of shares held in both the stock (SPY) and bond (AGG) portion of the portfolio.
spy_shares = 110
agg_shares = 200
```

#### Step 1: In the `Starter_Code` folder, create an environment file (`.env`) to store the values of your Alpaca API key and Alpaca secret key.

#### Step 2: Set the variables for the Alpaca API and secret keys. Using the Alpaca SDK, create the Alpaca `tradeapi.REST` object. In this object, include the parameters for the Alpaca API key, the secret key, and the version number.

```python
# Setting the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')

# Create the Alpaca tradeapi.REST object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = 'v2')
```

#### Step 3: Set the following parameters for the Alpaca API call:

- `tickers`: Use the tickers for the member’s stock and bond holdings.

- `timeframe`: Use a time frame of one day.

- `start_date` and `end_date`: Use the same date for these parameters, and format them with the date of the previous weekday (or `2020-08-07`). This is because you want the one closing price for the most-recent trading day.


```python
# Set the tickers for both SPY and AGG
tickers = ['AGG', 'SPY']

# Seting the timeframe to 1D 
timeframe = '1D'


# Seting the start and end date at the dates in isoformat 

start_date = pd.Timestamp('2021-11-29', tz = 'America/New_York').isoformat()
end_date = pd.Timestamp('2021-11-29', tz = 'America/New_York').isoformat()
```

#### Step 4: Get the current closing prices for `SPY` and `AGG` by using the Alpaca `get_barset` function. Format the response as a Pandas DataFrame by including the `df` property at the end of the `get_barset` function.

```python
# Use the Alpaca get_barset function to get current closing prices the portfolio and create a dataframe

spy_agg_df = alpaca.get_barset(
    tickers,
    timeframe,
    start = start_date,
    end = end_date
).df

# Reviewing the first 5 rows of the Alpaca DataFrame
spy_agg_df.head()
```

#### Step 5: Navigating the Alpaca response DataFrame, select the `SPY` and `AGG` closing prices, and store them as variables.

```python
# Accessing the closing price for AGG from the Alpaca DataFrame

agg_close_price = spy_agg_df['AGG']['close']
agg_close_price = float(agg_close_price)

# Print the AGG closing price
print(agg_close_price)
```

```python
# Accessing the closing price for SPY from the Alpaca DataFrame

spy_close_price =spy_agg_df['SPY']['close']
spy_close_price = float(spy_close_price)
print(spy_close_price)
```

#### Step 6: Calculate the value, in US dollars, of the current amount of shares in each of the stock and bond portions of the portfolio, and print the results.

```python
# Calculating the current value of the bond portion of the portfolio
agg_value = agg_close_price * agg_shares

# Printing the current value of the bond portfolio
print(agg_value)
```

```python
# Calculating the current value of the stock portion of the portfolio
spy_value = spy_close_price * spy_shares

# Printing the current value of the stock portfolio
print(spy_value)
```

```python
# Calculating the total value of the stock and bond portfolio
total_stocks_bonds = spy_value + agg_value

# Printing the current balance of the stock and bond portfolio
print(total_stocks_bonds)
```

```python
# Calculate the total value of the member's entire savings portfolio including the crypto wallet

total_portfolio = total_crypto_wallet + total_stocks_bonds

# Printing the entire portfolio and wallet balance
print(total_portfolio)
```

In this section, you’ll use the MCForecastTools library to create a Monte Carlo simulation for the member’s savings portfolio. To do this, complete the following steps:

1. Make an API call via the Alpaca SDK to get 3 years of historical closing prices for a traditional 60/40 portfolio split: 60% stocks (SPY) and 40% bonds (AGG).

#### Step 1: Make an API call via the Alpaca SDK to get 3 years of historical closing prices for a traditional 60/40 portfolio split: 60% stocks (SPY) and 40% bonds (AGG).

```python
# Setting start date 3 years prior to today 
start_date = pd.Timestamp('2018-11-29', tz = 'America/New_York').isoformat()
end_date = pd.Timestamp('2021-11-29', tz = 'America/New_York').isoformat()
# Seting number of rows to 1000 to retrieve the maximum amount of rows
limit_rows = 1000
```

```python
# Using the Alpaca get_barset function to make the API call to get the 3 years worth of pricing data

spy_agg_3yr = alpaca.get_barset(
    tickers,
    timeframe,
    start = start_date,
    end = end_date,
    limit = limit_rows
).df


# Displaying both the first and last five rows of the DataFrame
display(spy_agg_3yr.head())
display(spy_agg_3yr.tail())
```

#### Step 2: Run a Monte Carlo simulation of 500 samples and 30 years for the 60/40 portfolio, and then plot the results.

```python
# Configuring the Monte Carlo simulation to forecast 30 years of cumulative returns with a 40/60 split AGG/SPY

agg_spy_30yr = MCSimulation(
  portfolio_data = spy_agg_3yr,
  weights = [.40,.60],
  num_simulation = 500,
  num_trading_days = 252*30
)


# Reviewing the simulation input data
agg_spy_30yr.portfolio_data
```

```python
# Running the Monte Carlo simulation to forecast 30 years of cumulative returns
agg_spy_30yr.calc_cumulative_return()
```

```python
# Visualizing the 30-year Monte Carlo simulation 

agg_spy_30yr.plot_simulation()
```

#### Step 3: Plot the probability distribution of the Monte Carlo simulation.

```python
# Visualize the probability distribution of the 30-year Monte Carlo simulation 

agg_spy_30yr.plot_distribution()
```

#### Step 4: Generate the summary statistics for the Monte Carlo simulation.

```python
# Generating summary statistics from the 30-year Monte Carlo simulation results

mc_30yr_summary = agg_spy_30yr.summarize_cumulative_return()


# Reviewing the 30-year Monte Carlo summary statistics
display(mc_30yr_summary)
```

### Analyze the Retirement Portfolio Forecasts

Using the current value of only the stock and bond portion of the member's portfolio and the summary statistics that you generated from the Monte Carlo simulation, answer the following question in your Jupyter notebook:

-  What are the lower and upper bounds for the expected value of the portfolio with a 95% confidence interval?


```python
# Printing the current balance of the stock and bond portion of the members portfolio
print(total_stocks_bonds)
```

```python
# Using the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes for the current stock/bond portfolio
ci_lower_thirty_cumulative_return = total_stocks_bonds * mc_30yr_summary.iloc[8]
ci_upper_thirty_cumulative_return = total_stocks_bonds * mc_30yr_summary.iloc[9]

# Printing the results of your calculations
print(ci_lower_thirty_cumulative_return)
print(ci_upper_thirty_cumulative_return)
```

### Forecast Cumulative Returns in 10 Years

The CTO of the credit union is impressed with your work on these planning tools but wonders if 30 years is a long time to wait until retirement. So, your next task is to adjust the retirement portfolio and run a new Monte Carlo simulation to find out if the changes will allow members to retire earlier.

For this new Monte Carlo simulation, do the following: 

- Forecast the cumulative returns for 10 years from now. Because of the shortened investment horizon (30 years to 10 years), the portfolio needs to invest more heavily in the riskier asset&mdash;that is, stock&mdash;to help accumulate wealth for retirement. 

- Adjust the weights of the retirement portfolio so that the composition for the Monte Carlo simulation consists of 20% bonds and 80% stocks. 

- Run the simulation over 500 samples, and use the same data that the API call to Alpaca generated.

- Based on the new Monte Carlo simulation, answer the following questions in your Jupyter notebook:

    - Using the current value of only the stock and bond portion of the member's portfolio and the summary statistics that you generated from the new Monte Carlo simulation, what are the lower and upper bounds for the expected value of the portfolio (with the new weights) with a 95% confidence interval?

    - Will weighting the portfolio more heavily toward stocks allow the credit union members to retire after only 10 years?


```python
# Configure a Monte Carlo simulation to forecast 10 years cumulative returns
# The weights should be split 20% to AGG and 80% to SPY.

agg_spy_10yr = MCSimulation(
  portfolio_data = spy_agg_3yr,
  weights = [.20,.80],
  num_simulation = 500,
  num_trading_days = 252*10
)

# Review the simulation input data
agg_spy_10yr.portfolio_data
```

```python
# Running the Monte Carlo simulation to forecast 10 years cumulative returns
agg_spy_10yr.calc_cumulative_return()
```

```python
# Visualize the 10-year Monte Carlo simulation by creating an
# overlay line plot
agg_spy_10yr.plot_simulation()
```

```python
# Visualize the probability distribution of the 10-year Monte Carlo simulation 
# by plotting a histogram
agg_spy_10yr.plot_distribution()
```

```python
# Generate summary statistics from the 10-year Monte Carlo simulation results
# Save the results as a variable
mc_10yr_summary = agg_spy_10yr.summarize_cumulative_return()


# Review the 10-year Monte Carlo summary statistics
display(mc_10yr_summary)
```
